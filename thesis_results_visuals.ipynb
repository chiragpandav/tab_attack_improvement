```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv(r"D:\Data\Download\python_projects\tableak_adult_gen\all_exp_result\model_utilities_regular.csv")
data = df[(df['model_type'].isin(['normal', 'dp'])) & (df['dataset'] == 'bank')]
grouped = data.groupby(['model_type'])[['TPR', 'Accuracy']].mean()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

model_types = ['normal', 'dp']
tpr_values = [grouped.loc[mt, 'TPR'] * 100 for mt in model_types]
acc_values = [grouped.loc[mt, 'Accuracy'] * 100 for mt in model_types]

colors = ['#2E86AB', '#A23B72']
labels = ['Normal', 'DP']
bars1 = ax1.bar(model_types, tpr_values, color=colors, width=0.6, label=labels)
for i, v in enumerate(tpr_values):
    ax1.text(i, v + 1, f'{v:.1f}%', ha='center', va='bottom', fontweight='bold')
ax1.set_ylabel('TPR (%)', fontsize=12, fontweight='bold')
ax1.set_title('TPR Comparison', fontsize=14, fontweight='bold', pad=20)
ax1.set_ylim(0, 100)
ax1.legend(title='Model Type', title_fontsize=11, fontsize=10, frameon=True, fancybox=True, shadow=True)
ax1.grid(True, alpha=0.3, linestyle='--')
ax1.tick_params(axis='both', which='major', labelsize=10)

bars2 = ax2.bar(model_types, acc_values, color=colors, width=0.6, label=labels)
for i, v in enumerate(acc_values):
    ax2.text(i, v + 1, f'{v:.1f}%', ha='center', va='bottom', fontweight='bold')
ax2.set_ylabel('Accuracy (%)', fontsize=12, fontweight='bold')
ax2.set_title('Accuracy Comparison', fontsize=14, fontweight='bold', pad=20)
ax2.set_ylim(0, 100)
ax2.legend(title='Model Type', title_fontsize=11, fontsize=10, frameon=True, fancybox=True, shadow=True)
ax2.grid(True, alpha=0.3, linestyle='--')
ax2.tick_params(axis='both', which='major', labelsize=10)

plt.tight_layout()
plt.show()
```
