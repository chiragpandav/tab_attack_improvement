```python
group_pairs = [
    ("tabl_REG_NPR", "tabl_REG_PRE"),
    ("tabl_QAT_NPR", "tabl_QAT_PRE")
]

for pair in group_pairs:
    combined_data = []
    for group in pair:
        norm = globals()[f'PUMS_{group.replace("tabl", "tabl_norm")}']
        dp = globals()[f'PUMS_{group.replace("tabl", "tabl_dp")}']
        fair = globals()[f'PUMS_{group.replace("tabl", "tabl_fair")}']
        fair_1 = globals()[f'PUMS_{group.replace("tabl", "tabl_fair")}_1']
        data = pd.DataFrame({
            'Normal': norm["overall"],
            'DP': dp["overall"],
            'Fair': fair["overall"],
            'FairDp': fair_1["overall"]
        })
        data_melted = pd.melt(data, var_name='Model', value_name='Values')
        data_melted['Group'] = group
        combined_data.append(data_melted)
    
    plot_data = pd.concat(combined_data, ignore_index=True)
    plt.figure(figsize=(12, 6))
    ax = sns.boxplot(x='Model', y='Values', hue='Group', data=plot_data)
    
    # Custom titles
    custom_titles = {
        ("tabl_REG_NPR", "tabl_REG_PRE"): "Regular Comparison",
        ("tabl_QAT_NPR", "tabl_QAT_PRE"): "QAT Comparison"
    }
    plt.title(custom_titles.get(pair, f'Comparison: {pair[0]} & {pair[1]}'))
    
    # Get current handles and labels, then replace labels
    handles, labels = ax.get_legend_handles_labels()
    custom_legend_labels = {
        ("tabl_REG_NPR", "tabl_REG_PRE"): ["Tableak", "Tableak_WithPreTrained"],
        ("tabl_QAT_NPR", "tabl_QAT_PRE"): ["Tableak_QAT", "Tableak_QAT_WithPreTrained"]
    }
    new_labels = custom_legend_labels.get(pair, pair)
    plt.legend(handles, new_labels, title='Attack Mode')
    
    plt.xlabel('Training Mode')
    plt.ylabel('Reconstruction %')
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

columns_to_plot = ["SEX_1", "SEX_2", "RAC1P_1", "RAC1P_2", "SEX_1_RAC1P_1", "SEX_1_RAC1P_2", "SEX_2_RAC1P_1", "SEX_2_RAC1P_2"]

column_labels = {"SEX_1": "Male", "SEX_2": "Female", "RAC1P_1": "White", "RAC1P_2": "Black", "SEX_1_RAC1P_1": "Male & White", "SEX_1_RAC1P_2": "Male & Black", "SEX_2_RAC1P_1": "Female & White", "SEX_2_RAC1P_2": "Female & Black"}

custom_legend_labels = {
    ("tabl_REG_NPR", "tabl_REG_PRE"): ["PUMS", "PUMS (WithPreTrained)"],
    ("tabl_QAT_NPR", "tabl_QAT_PRE"): ["PUMS (QAT)", "PUMS (QAT+WithPreTrained)"]
}


group_pairs = [("tabl_REG_NPR", "tabl_REG_PRE"), ("tabl_QAT_NPR", "tabl_QAT_PRE")]

colors = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D']
palette = sns.color_palette(colors)

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette(palette)

for group_pair in group_pairs:
    for i, group in enumerate(group_pair):
        norm = globals()[f'PUMS_{group.replace("tabl", "tabl_norm")}']
        dp = globals()[f'PUMS_{group.replace("tabl", "tabl_dp")}']
        fair = globals()[f'PUMS_{group.replace("tabl", "tabl_fair")}']
        fair_1 = globals()[f'PUMS_{group.replace("tabl", "tabl_fair")}_1']
        df_list = []
        
        for model_name, df in zip(['Normal', 'DP', 'Fair', 'FairDp'], [norm, dp, fair, fair_1]):
            melted = df[columns_to_plot].melt(var_name='Category', value_name='Values')
            melted['Model'] = model_name
            melted['Category'] = melted['Category'].map(column_labels)
            df_list.append(melted)

        plot_data = pd.concat(df_list, ignore_index=True)
        title = custom_legend_labels[group_pair][i]

        fig, ax = plt.subplots(figsize=(14, 6))
        box_plot = sns.boxplot(x='Model', y='Values', hue='Category', data=plot_data, ax=ax, palette='Set2')
        
        ax.set_title(title, fontsize=16, fontweight='bold', pad=20)
        ax.set_xlabel('Training Mode', fontsize=12, fontweight='bold')
        ax.set_ylabel('Reconstruction %', fontsize=12, fontweight='bold')
        ax.grid(True, alpha=0.3)
        
        handles, labels = ax.get_legend_handles_labels()
        ax.legend(handles, labels, title='Demographic Group', title_fontsize=11, fontsize=10, bbox_to_anchor=(1.02, 1), loc='upper left', frameon=True, fancybox=True, shadow=True)
        
        plt.tight_layout()
        plt.show()
```
